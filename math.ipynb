```python
digits = 9
sys = {"ind": {"TC": 0, "C": 1, "TL": 2, "L": 3, "TTh": 4, "Th": 5, "H": 6, "T": 7, "U": 8},
       "int": {"HM": 0, "TM": 1, "M": 2, "HTh": 3, "TTh": 4, "Th": 5, "H": 6, "T": 7, "U": 8}}
sys_inv = {"ind": ["TC", "C", "TL", "L", "TTh", "Th", "H", "T", "U"],
           "int": ["HM", "TM", "M", "HTh", "TTh", "Th", "H", "T", "U"]}
tab_space = 8
```

```python
class Number():
    def __init__(self, value):
        self._value = [0,
                       (value // 100000000) % 10,
                       (value // 10000000) % 10,
                       (value // 1000000) % 10,
                       (value // 100000) % 10,
                       (value // 10000) % 10,
                       (value // 1000) % 10,
                       (value // 100) % 10,
                       (value // 10) % 10,
                       value % 10]
        return
    
    def get_value(self, digit):
        return self._value[digit]
    
    def set_value(self, digit, value):
        self._value[digit] = value
        return
```

```python
def add(number_0, number_1, system):
    carry = Number(0)
    answer = Number(0)
    steps = digits + 1
    while number_0.get_value(digits + 3 - steps) == 0 and number_1.get_value(digits + 3 - steps) == 0:
        steps -= 1
    for step in range(steps):
        print("Step ", step, "\n\n", end="", sep="")
        print("%-16s" % "Places:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % sys_inv[system][digit], end="", sep="")
        print()
        print()
        print("%-16s" % "Carry:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % carry.get_value(digit + 1), end="", sep="")
        print()
        print()
        print("%-16s" % "Number 1:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % number_0.get_value(digit + 1), end="", sep="")
        print()
        print("%-12s" % "Number 2:", "%4s" % "+", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % number_1.get_value(digit + 1), end="", sep="")
        print()
        dashes = ""
        for i in range(16 + tab_space * digits):
            dashes += "-"
        print(("%-" + str(16 + tab_space * digits) + "s") % dashes)
        print("%-16s" % "Answer:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % answer.get_value(digit + 1), end="", sep="")
        answer.set_value(digits - step, (carry.get_value(digits - step) + number_0.get_value(digits - step) + number_1.get_value(digits - step)) % 10)
        carry.set_value(digits - step - 1, (carry.get_value(digits - step) + number_0.get_value(digits - step) + number_1.get_value(digits - step)) // 10)
        print("\n\nAdding digits at ", sys_inv[system][digits - step - 1], " place.\n\n", sep="", end="")
        print(carry.get_value(digits - step), " + ", number_0.get_value(digits - step), " + " ,number_1.get_value(digits - step), " = ", carry.get_value(digits - step - 1) * 10 + answer.get_value(digits - step), sep="", end="")
        print("\n\nHence, digit in answer at ", sys_inv[system][digits - step - 1], " place will be ", answer.get_value(digits - step), ".\n", sep="", end="")
        print("And digit in carry at ", sys_inv[system][digits - step - 2], " place will be ", carry.get_value(digits - step - 1), ".\n\n\n", sep="", end="")
    return
```

```python
add(Number(7044), Number(5591), "ind")
```

```python
def subtract(number_0, number_1, system):
    answer = Number(0)
    steps = digits + 1
    while number_0.get_value(digits + 2 - steps) == 0 and number_1.get_value(digits + 2 - steps) == 0:
        steps -= 1
    for step in range(steps):
        print("Step ", step, "\n\n", end="", sep="")
        print("%-16s" % "Places:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % sys_inv[system][digit], end="", sep="")
        print()
        print()
        print("%-16s" % "Number 1:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % number_0.get_value(digit + 1), end="", sep="")
        print()
        print("%-12s" % "Number 2:", "%4s" % "-", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % number_1.get_value(digit + 1), end="", sep="")
        print()
        dashes = ""
        for i in range(16 + tab_space * digits):
            dashes += "-"
        print(("%-" + str(16 + tab_space * digits) + "s") % dashes)
        print("%-16s" % "Answer:", end="", sep="")
        for digit in range(digits):
            print(("%" + str(tab_space) + "s") % answer.get_value(digit + 1), end="", sep="")
        if number_0.get_value(digits - step) >= number_1.get_value(digits - step):
            print("\n\nSubtracting digits at ", sys_inv[system][digits - step - 1], " place.\n\n", end="", sep="")
            print(number_0.get_value(digits - step), " - ", number_1.get_value(digits - step), " = ", answer.get_value(digits - step), "\n", end="", sep="")
        else:
            print("\n\nAs digit at ", sys_inv[system][digits - step - 1], " place of number 1 is less than that of number 2, we need to (???).\n\n", end="", sep="")
        answer.set_value(digits - step, number_0.get_value(digits - step) - number_1.get_value(digits - step))
        if answer.get_value(digits - step) < 0:
            answer.set_value(digits - step, answer.get_value(digits - step) + 10)
            number_0.set_value(digits - step, number_0.get_value(digits - step) + 10)
            number_0.set_value(digits - step - 1, number_0.get_value(digits - step - 1) - 1)
            i = 0
            while number_0.get_value(digits - step - 1 - i) < 0:
                number_0.set_value(digits - step - 1 - i, number_0.get_value(digits - step - 1 - i) + 10)
                number_0.set_value(digits - step - 2 - i, number_0.get_value(digits - step - 2 - i) - 1)
                i += 1
            while i > 0:
                print("At ", sys_inv[system][digits - step - 2 - i], " place, ", number_0.get_value(digits - step - 1 - i) + 1, " - 1 = ", number_0.get_value(digits - step - 1 - i), ".\n", end="", sep="")
                print("At ", sys_inv[system][digits - step - 1 - i], " place, ", number_0.get_value(digits - step - i) - 9, " + 10 = ", number_0.get_value(digits - step - i) + 1, ".\n", end="", sep="")
                i -= 1
            print("At ", sys_inv[system][digits - step - 2], " place, ", number_0.get_value(digits - step - 1) + 1, " - 1 = ", number_0.get_value(digits - step - 1), ".\n", end="", sep="")
            print("At ", sys_inv[system][digits - step - 1], " place, ", number_0.get_value(digits - step) - 10, " + 10 = ", number_0.get_value(digits - step), ".\n", end="", sep="")
        '''print("\n\nAdding digits at ", sys_inv[system][digits - step - 1], " place.\n\n", sep="", end="")
        print(carry.get_value(digits - step), " + ", number_0.get_value(digits - step), " + " ,number_1.get_value(digits - step), " = ", carry.get_value(digits - step - 1) * 10 + answer.get_value(digits - step), sep="", end="")
        print("\n\nHence, digit in answer at ", sys_inv[system][digits - step - 1], " place will be ", answer.get_value(digits - step), ".\n", sep="", end="")
        print("And digit in carry at ", sys_inv[system][digits - step - 2], " place will be ", carry.get_value(digits - step - 1), ".\n\n\n", sep="", end="")
        '''
        print()
        print()
    return
```

```python
def round_to_digit(number_0, digit):
    print("%-16s" % "", "%-4s" % "TC", "%-4s" % "C", "%-4s" % "TL", "%-4s" % "L", "%-4s" % "TTh", "%-4s" % "Th", "%-4s" % "H", "%-4s" % "T", "%-4s" % "U", sep="")
    print()
    print("%-16s" % "Number:", "%-4s" % number_0.get_value(0), "%-4s" % number_0.get_value(1), "%-4s" % number_0.get_value(2), "%-4s" % number_0.get_value(3), "%-4s" % number_0.get_value(4), "%-4s" % number_0.get_value(5), "%-4s" % number_0.get_value(6), "%-4s" % number_0.get_value(7), "%-4s" % number_0.get_value(8), sep="")
    print(("%" + str(16 + 4 * (digit + 1)) + "s") % "Do Not Change <= | ", end="", sep="")
    if number_0.get_value(digit + 1) < 5:
        print("%-4s" % "<5", end="", sep="")
    else:
        print("%-4s" % ">=5", end="", sep="")
        number_0.set_value(digit, number_0.get_value(digit) + 1)
        i = 0
        while number_0.get_value(digit - i) >= 10:
            number_0.set_value(digit - i, number_0.get_value(digit - i) - 10)
            number_0.set_value(digit - i - 1, number_0.get_value(digit - i - 1) + 1)
            i += 1
    if digit != 8:
        print(" => Change to 0s", end="", sep="")
    for i in range(digits - digit - 1):
        number_0.set_value(digit + i + 1, 0)
    print()
    print()
    print()
    print("%-16s" % "", "%-4s" % "TC", "%-4s" % "C", "%-4s" % "TL", "%-4s" % "L", "%-4s" % "TTh", "%-4s" % "Th", "%-4s" % "H", "%-4s" % "T", "%-4s" % "U", sep="")
    print()
    print("%-16s" % "Number:", "%-4s" % number_0.get_value(0), "%-4s" % number_0.get_value(1), "%-4s" % number_0.get_value(2), "%-4s" % number_0.get_value(3), "%-4s" % number_0.get_value(4), "%-4s" % number_0.get_value(5), "%-4s" % number_0.get_value(6), "%-4s" % number_0.get_value(7), "%-4s" % number_0.get_value(8), sep="")
    print()
    print()
    return
```
